## 使用 tf.keras 搭建神经网络

使用 tf.keras 搭建神经网络有着基本固定的套路，并且可以极大的简化搭建神经网络所用的代码量。

下面来逐步讲解使用 tf.keras 搭建神经网络的六个基本步骤：
 1. import 相关模块
 2. 指定输入网络的训练集和测试集
 3. 使用 tf.keras.models.Sequential 来逐层搭建网络结构模型 model
 4. 使用 model.compile() 来配置训练方法
 5. 使用 model.fit() 来执行训练过程
 6. 使用 model.summary() 来打印网络结构

### 3. 逐层搭建网络结构

这一步涉及到的常用的 tf.keras API 如下所示：

#### 3.1 tf.keras.models.Squential()

Sequential 函数是一个容器，描述了神经网络的网络结构，在 Sequential 函数的输入参数中描述从输入层到输出层的网络结构。 

在构建网络结构的步骤中，自然少不了对各种类型的网络层进行表示，常用的有下面几种类型

##### 3.1.1 拉直层

`tf.keras.layers.Flatten()` 

拉直层可以变换张量的尺寸，把输入特征拉直为一维数组，是不含计算参数的层。

##### 3.1.2 全连接层：

`tf.keras.layers.Dense(神经元个数, activation="激活函数", kernel_regularizer="正则化方法")`

其中：
 + activation 可选参数有 `"relu"`、`"softmax"`、`"sigmoid"`、`"tanh"` 等
 + kernel_regularizer 可选参数有 `tf.keras.regularizers.l1()`、`tf.keras.regularizers.l2()`
 
##### 3.1.1 卷积层：

`tf.keras.layers.Conv2D(filter, kernel_size, strides, padding)`

其中：
 + filter 表示卷积核个数
 + kernel_size 表示卷积核尺寸
 + strides 表示卷积步长或者步幅，是应用滤波器的位置间隔
 + padding 表示填充，即向输入数据的周围填入固定数据，可选参数为 `"valid"` 或者 `"same"`

##### 3.1.1 LSTM层：

`tf.keras.layers.LSTM()`

### 4. 配置训练方法

这一步涉及到的函数主要是 `model.compile(optimizer, loss, metrics)`。用于配置神经网络的训练方法，告知训练时使用的优化器、损失函数和准确率评测标准

#### 4.1 optimzier 

即优化器，参数值可以使用字符串的形式给出优化器名字，也可以使用函数形式，使用函数形式时，可以同时设置优化器的学习率、动量和超参数等。可用选项简单介绍如下：

 +`"sgd"` 或者 `tf.optimizers.SGD( lr=学习率. decay=学习率衰减率, momentum=动量参数)`
 + `"adagrad"` 或者 `tf.keras.optimizers.Adagrad(lr=学习率, decay=学习率衰减率)`
 + `"adadelta"` 或者 `tf.keras.optimizers.Adadelta(lr=学习率, decay=学习率衰减率)`
 +`"adam"` 或者 `tf.keras.optimizers.Adam (lr=学习率, decay=学习率衰减率)`
 
#### 4.2 loss

即损失函数，同样也可以给出字符串形式的损失函数名字，或者函数形式。可用选项简单介绍如下：
 + `"mse"` 或者 `tf.keras.losses.MeanSquaredError()`
 + `"sparse_categorical_crossentropy"` 或者 `tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)`

对参数 `from_logits` 的理解：有些神经网络的输出是经过了 softmax 等函数的概率分布，有些则不经概率分布直接输出，而参数 `from_logits` 就是在询问是否是原始输出，也就是没有经过概率分布的输出，取 False 时表示转化为概率分布，取 True 时表示没有转化为概率分布，直接输出。 

#### 4.3 metrics

即标注网络评测指标。可用选项简单介绍如下：(y_ 为真实值，y 为预测值)

 + `"accuracy"`表示真实值 y_ 和输出预测值 y 都是数值形式，如 `y_=[1]`、`y=[1]` 
 + `"categorical_accuracy"`表示真实值 y_ 和输出预测值 y 都是独热码或概率分布表示，如 `y_=[0, 1, 0]`、`y=[0.256, 0.695, 0.048]`
 + `"sparse_ categorical_accuracy"`表示真实值 y_  是以数值形式给出，输出预测值 y 是以独热码或概率分布形式给出。如 `y_=[1]`、`y=[0.256, 0.695, 0.048]`

### 5. 执行训练过程

这一步使用的 API 主要为 

```python
model.fit(训练集的输入特征， 训练集的标签， 
        batch_size=每次喂入神经网络的样本数, 
        epochs=迭代数据集的次数,  
        validation_data=(测试集的输入特征，测试集的标签)， 
        validataion_split=从训练集中划分多少比例的数据作为测试集， 
        validation_freq=多少次epoch测试一次) 
```

`validation_data` 和 `validataion_split` 这两个参数二者选其一即可。

### 6. 打印网络结构

使用 `model.summary()` 函数来打印网络结构和参数统计

### 实践：使用 tf.keras 重新实现神经网络对 iris 数据集的分类

### 1. import 相关模块

````python
import tensorflow as tf
import numpy as np
from sklearn import datasets # 为了使用 iris 数据集
````

### 2. 指定输入网络的训练集和测试集

````python
# 导入数据集
x_train = datasets.load_iris().data
y_train = datasets.load_iris().target

# 乱序数据集
np.random.seed(666)
np.random.shuffle(x_train)
np.random.seed(666)
np.random.shuffle(y_train)
````

###  3. 使用 tf.keras.models.Sequential 来逐层搭建网络结构模型 model

````python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(3, activation="softmax",kernel_regularizer=tf.keras.regularizers.l2())
])
````

###  4. 使用 model.compile() 来配置训练方法


````python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])
````

### 5. 使用 model.fit() 来执行训练过程

````python
model.fit(x_train, y_train, batch_size=32, epochs=500, validation_freq=20, validation_split=0.2)
````

### 6. 使用 model.summary() 来打印网络结构

````python
model.summary()
````
